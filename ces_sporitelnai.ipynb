# LEETCODE - reverse words in a string

obrazek.png

```python
sentence = "the sky is si blue"
# rozdělěení stringu
words_list = sentence.split()
# nové pole, kde bude obrácená věta
worlds_list_rev=[]
# největší číslo
highest_numb = len(words_list)
# cyklus, kde prohodíem slova
for item in words_list:
    highest_numb=highest_numb-1
    worlds_list_rev.append(words_list[highest_numb])
print(worlds_list_rev)
```

# LEET CODE - rotate array

obrazek.png

```python
nums = [1,2,3,4,5,6,7]
k = 2

#vytvoření nového pole, kam to budem vkládat
rotated_nums = [0] * len(nums)

#
for i in range(len(nums)):
    # vypočtení nového indexu, abychom změnili pořadí
    new_index = (i + k) % len(nums)
    rotated_nums[new_index] = nums[i]


print(rotated_nums)
```

# Code Wars Multiples of 3 or 5

obrazek.png

```python
# hlavní proměnná
num = 15


array = []
final_count = 0
count = 0

#cyklus, který nám z hodnoty co zadáme vytvoří pole čísel, jelikož zkoumáme POD tak mažeme nejvyšší číslici
while count != num-1:
    count += 1
    array.append(count)
    print(count)
print(array)

# každé číslo v poli otestujem, jestli ej dělitelné 3 nebo 5
for item in array:
    result_1 = item % 3
    result_2 = item % 5
    #print(result_1)
    #print(result_2)
    if result_1==0 or result_2 == 0:
        print("ANO",item)
        final_count = final_count + item
print(f"vysledek: {final_count}")
```

# HACKERRANK - přestupný rok

obrazek.png

```python
def is_leap(year):
    leap = False

    if year % 4 == 0:
        if year % 100 != 0 or year % 400 == 0:
            leap = True


    return leap

year = int(1900)
is_leap(year)
```
